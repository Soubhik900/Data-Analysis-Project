# Removing Null Values

```python
median1=data['Enroll'].median()
median2=data['S.F.Ratio'].median()

data["Enroll"].replace(np.nan,median1,inplace=True)
data["S.F.Ratio"].replace(np.nan,median2,inplace=True)
```

# Removing Outliers

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range    

for i in data.columns:
    if data[i].dtype !='object':
        lr,ur=remove_outlier(data[i])
        data[i]=np.where(data[i]>ur,ur,data[i])
        data[i]=np.where(data[i]<lr,lr,data[i])
```

# 9.	Overall Insights


**•	Which factors (applications, acceptance rate, enrollment, academic excellence, costs, faculty qualifications, student/faculty ratio, alumni donations, expenditures) are most strongly associated with higher graduation rates?**


```python
correlations = {
    'Applications': data['Apps'].corr(data['Grad.Rate']),
    'Acceptance Rate': (data['Accept'] / data['Apps']).corr(data['Grad.Rate']),
    'Enrollment Rate': (data['Enroll'] / data['Accept']).corr(data['Grad.Rate']),
    'Top 10%': data['Top10perc'].corr(data['Grad.Rate']),
    'Top 25%': data['Top25perc'].corr(data['Grad.Rate']),
    'Room and Board': data['Room.Board'].corr(data['Grad.Rate']),
    'Books': data['Books'].corr(data['Grad.Rate']),
    'Personal Spending': data['Personal'].corr(data['Grad.Rate']),
    'PhD Faculty': data['PhD'].corr(data['Grad.Rate']),
    'Terminal Faculty': data['Terminal'].corr(data['Grad.Rate']),
    'Student/Faculty Ratio': data['S.F.Ratio'].corr(data['Grad.Rate']),
    'Alumni Donations': data['perc.alumni'].corr(data['Grad.Rate']),
    'Instructional Expenditure': data['Expend'].corr(data['Grad.Rate'])
}

for factor, correlation in correlations.items():
    print(f"Correlation between {factor} and graduation rate: {correlation:.2f}")

most_strongly_associated = max(correlations, key=correlations.get)
print(f"\nThe factor most strongly associated with higher graduation rates is {most_strongly_associated} with a correlation coefficient of {correlations[most_strongly_associated]:.2f}.")
```
